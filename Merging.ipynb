```python
import pandas as pd
from glob import glob
```

```python
data_files = (glob('20*/cq*/endfile.csv'))
data_files
```

```python
for i in range(len(data_files)):
    df = pd.read_csv(data_files[i])
    df.columns = ['closedDate', 'dealvalue','transaction_id','taget_id','target_companytype','buyer_id']
    df.to_csv(data_files[i])
```

```python
mergeddata = pd.concat(pd.read_csv(datafile).assign(sourcefilename = datafile)
                      for datafile in data_files)
```

```python
for i in range(2005,2023):
    data_files = (glob(str(i)+'/cq*/endfile.csv'))
    for i in range(len(data_files)):
        df = pd.read_csv(data_files[i])
        df.columns = ['closedDate', 'dealvalue','transaction_id','taget_id','target_companytype','buyer_id']
        df.to_csv(data_files[i])
    mergeddata = pd.concat(pd.read_csv(datafile).assign(sourcefilename = datafile)
                          for datafile in data_files)
```

```python
mergeddata.to_csv('final.csv')
```
